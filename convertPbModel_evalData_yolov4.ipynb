```python
#install pre-req
!pip install pillow numpy -q

from PIL import Image
import numpy as np

def pre_process_image(imagePath, img_height=416):
    # Model input format
    n, c, h, w = [1, 3, img_height, img_height]
    image = Image.open(imagePath)
    image = image.resize((h, w), resample=Image.BILINEAR)

    # Normalize to keep data between 0 - 1
    image = np.array(image) 

    # Change data layout from HWC to CHW
    image = image.transpose((2, 0, 1))
    input_image = image.reshape((n, c, h, w))

    return input_image
```

```python
from openvino.inference_engine import IECore
import cv2
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf

model_xml = f'{ir_out_dir}/{ir_name}.xml'
model_bin = f'{ir_out_dir}/{ir_name}.bin'

# Load network to the plugin
ie = IECore()
net = ie.read_network(model=model_xml, weights=model_bin)
exec_net = ie.load_network(network=net, device_name="CPU")
del net

input_layer = next(iter(exec_net.input_info))
output_layer = next(iter(exec_net.outputs))

# (1, 18, 13, 13) conv17
# (1, 18, 26, 26) conv20

output_layer_17 = 'StatefulPartitionedCall/model/conv2d_17/BiasAdd/Add'
output_layer_20 = 'StatefulPartitionedCall/model/conv2d_20/BiasAdd/Add'

# Run the Inference on the Input image...
inp_img_url = "drive/MyDrive/OpenCV-AI-Models/yolov4TinyStrawberry/"
inp_file_name = 'rgb-00003-16160356284338.jpeg'


#Download the image from the storage
# output = subprocess.check_output(f"curl {inp_img_url} --output {inp_file_name}", shell=True)

#Pre-process the image and get it ready for inference.
input_image = np.array(pre_process_image(inp_img_url+inp_file_name))

print(input_image.max())
# input_image = input_image / 255.
res = exec_net.infer(inputs={input_layer: input_image})

# (1, 18, 13, 13)
res17 = res[output_layer_17]

# (1, 18, 26, 26)
res20 = res[output_layer_20]

# aaa = res17[0, :, 0, 0]
# res17[0, 0, ...]
score = tf.nn.softmax(res17[0])
```

## Convert OpenVINO IR to Blob format 

```python
#CHOOSE the directory where you would like to save the blob.
# I use the same --output_dir as above for the IR conversion
blob_dir = f'/content/{ir_out_dir}/'

#Copy the path of your .xml and .bin files. For that, you can look at the IR
#conversion output cell, select and copy from:
#[SUCCESS] XML file and bin file paths.
#Or you can choose to compile other .xml .bin files from a different location
#
xmlfile = f'/content/{ir_out_dir}/{ir_name}.xml'
binfile = f'/content/{ir_out_dir}/{ir_name}.bin'

import requests
from io import StringIO
class Version:
    v2019_3 = '2019.R3'
    v2020_1 = '2020.1'
    v2020_2 = '2020.2'
    v2020_3 = '2020.3'
    v2020_4 = '2020.4'
    v2021_1 = '2021.1'
    v2021_2 = '2021.2'

url = "http://69.164.214.171:8084/compile"
params = {
    'version': Version.v2021_2
}
payload = {
    'name': f'{ir_name}',
    'myriad_shaves': 6
}
config = StringIO("""
task_type: detection
files:
  - name: FP16/yolov4TinyStrawberry.xml
    source:
      $type: http
      url: $REQUEST/yolov4TinyStrawberry.xml
  - name: FP16/flower.bin
    source:
      $type: http
      url: $REQUEST/yolov4TinyStrawberry.bin
framework: dldt
""")
files = {
    'config': config,
    'xml': open(xmlfile, 'rb'),
    'bin': open(binfile, 'rb')
}
response = requests.request("POST", url, data=payload, files=files, params=params)
print(response.headers.get('Content-Disposition'))
print(response.headers)
blobnameraw = response.headers.get('Content-Disposition')
blobname = blobnameraw[blobnameraw.find('='):][1:]
with open(blob_dir + blobname, 'wb') as f:
  f.write(response.content)
```
